```python
import matplotlib.pyplot as plt
import numpy as np

# Create data for the plots
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Initialize the figure and Axes objects
fig, axs = plt.subplots(2, 2, figsize=(10, 8))  # 2x2 grid of subplots

# Plot 1: Line plot with a legend
axs[0, 0].plot(x, y1, label='sin(x)', color='blue', linewidth=2)
axs[0, 0].set_title("Sine Wave")
axs[0, 0].set_xlabel("X-axis")
axs[0, 0].set_ylabel("Y-axis")
axs[0, 0].legend()
axs[0, 0].grid(True)

# Plot 2: Scatter plot with customization
scatter_x = np.random.uniform(0, 10, 50)
scatter_y = np.random.uniform(-1, 1, 50)
axs[0, 1].scatter(scatter_x, scatter_y, color='green', alpha=0.7, edgecolor='black')
axs[0, 1].set_title("Random Scatter")
axs[0, 1].set_xlabel("Random X")
axs[0, 1].set_ylabel("Random Y")
axs[0, 1].grid(True)

# Plot 3: Bar chart
bars = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]
axs[1, 0].bar(bars, values, color='orange', edgecolor='black')
axs[1, 0].set_title("Bar Chart")
axs[1, 0].set_xlabel("Categories")
axs[1, 0].set_ylabel("Values")

# Plot 4: Polar plot
axs[1, 1] = plt.subplot(2, 2, 4, polar=True)  # Polar subplot
theta = np.linspace(0, 2 * np.pi, 100)
r = 1 + np.sin(4 * theta)
axs[1, 1].plot(theta, r, color='purple')
axs[1, 1].set_title("Polar Plot", va='bottom')

# Adjust layout and show the figure
plt.tight_layout()
plt.show()
```
